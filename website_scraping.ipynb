Creating dataframes out of dictionaries using the Website identifier as a common key

```python
tweet = pd.DataFrame(list(twit.items()),columns = ['Website','Twitter'])
faceb = pd.DataFrame(list(fb.items()),columns = ['Website','Facebook'])
inst = pd.DataFrame(list(insta.items()),columns = ['Website','Instagram'])
ytube = pd.DataFrame(list(yout.items()),columns = ['Website','Youtube'])
linked = pd.DataFrame(list(linke.items()),columns = ['Website','LinkedIn'])
```

Merging the datasets with the initial dataset using inner and left join

```python
dfinal = sample.merge(tweet, on="Website",how="inner")
dfinal = dfinal.merge(faceb, on="Website",how="left")
dfinal = dfinal.merge(inst, on="Website",how="left")
dfinal = dfinal.merge(ytube, on="Website",how="left")
dfinal = dfinal.merge(linked, on="Website",how="left")
```
