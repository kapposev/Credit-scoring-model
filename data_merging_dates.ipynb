Dropping companies which havent provided financial informaiton for more than 36 months

```python
noactive=df[df['Company status']!='Active']
noactive['Latest accounts date']=pd.to_datetime(noactive['Latest accounts date'])
noactive['Status date']=pd.to_datetime(noactive['Status date'])
#dd=noactive[noactive['Latest accounts date']<pd.to_datetime('31/12/2017')]
noactive["diff_account_status"] = noactive['Status date'] - noactive['Latest accounts date']
noactive['diff_account_status']=noactive['diff_account_status'].astype('timedelta64[M]')
noactive=noactive[noactive['diff_account_status']<=36]
noactive
```

```python
noactive
```

```python
active=df[df['Company status']=='Active']
active['Latest accounts date']=pd.to_datetime(active['Latest accounts date'])
active["diff_from_now"] = pd.to_datetime('01/06/2020')-active['Latest accounts date']
active["diff_from_now"]=active["diff_from_now"].astype('timedelta64[M]')
active=active[active["diff_from_now"]<=36]
active
```

```python
active=active.rename(columns={'diff_from_now': 'diff'})
noactive=noactive.rename(columns={'diff_account_status': 'diff'})
```
