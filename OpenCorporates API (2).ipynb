```python
scottish=[]
for i in companies:
    for v in i["results"]["companies"]:
        if "SC" in v["company"]["company_number"]:
            scottish.append(v["company"])
        
```

```python
openl=[]
closed=[]
for i in scottish:
    if i["current_status"]=="Dissolved":
        closed.append(i)
    if i["current_status"]=="Active":
        openl.append(i)
print("closed companies:",len(closed))
print("open companies:",len(openl))   
```
