#  Building a fuction for claning the raw Twitter ID data

```python
def convert(row):
    #countl=countl + 1
    place=0
    if "home?status=" in row or "dotukdotcom" in row or 'intent/tweet?' in row or 'share' in row:
        return pd.np.nan
    elif "screen_name=" in row:
        place=row.index("screen_name=")+12
        return row[place:]
    elif "#" in row:
        place=row.index("#")
        return row[:place]
    elif "#!/" in row:
        place=3
        return row[place:]
    elif "/statuses" in row:
        place=row.index("/statuses")
        return row[:place]
    elif "/" in row:
        #print(i,"before")
        place=row.index("/")
        return row[:place]        
    elif "?" in row:
        place=row.index("?")
        return row[:place]
    else:
        return row
df['Twitter']=df['Twitter'].map(convert) 
df= df[df["Twitter"].notnull()]
```

Collecting information for each company that has a Twitter account and store it dictionaries

```python
count=0
followers={}
followed_accounts={}
lists_number={}
likes={}
Tweets={}
verification={}
date={}
for i in df['Twitter']:
    count=count+1
    try:
        user = api.get_user("{}".format(i)) # using tha API
        print(user.screen_name,count)
        followers[i]=user.followers_count #extracting the number of followers and store it in the dictionary
        followed_accounts[i]=user.friends_count
        lists_number[i]=user.listed_count
        likes[i]=user.favourites_count
        Tweets[i]=user.statuses_count
        verification[i]=user.verified
        date[i]=user.created_at
        #valid.append(i)
    except:
        print(i,"there is no information",count) # if there is no valid twitter account or it has colsed print this
```

Creating data frames out the dictionaries for each different social media platform

```python
follower_num = pd.DataFrame(list(followers.items()),columns = ['Twitter','Followers_number'])
tweets = pd.DataFrame(list(Tweets.items()),columns = ['Twitter','Tweets_number'])
followed_ones = pd.DataFrame(list(followed_accounts.items()),columns = ['Twitter','Accounts_followed'])
Likes = pd.DataFrame(list(likes.items()),columns = ['Twitter','Number_of_likes'])
verify = pd.DataFrame(list(verification.items()),columns = ['Twitter','Verification'])
lists_num = pd.DataFrame(list(lists_number.items()),columns = ['Twitter','Number_of_lists_included'])
dates_creation= pd.DataFrame(list(date.items()),columns = ['Twitter','Years of Twitter account'])
```

Merging the datasets using the Twitter ID as the common key with inner join

```python
dfinal = df.merge(follower_num,how="inner",on='Twitter')
dfinal=dfinal.merge(tweets,how="inner",on='Twitter')
dfinal=dfinal.merge(followed_ones,how="inner",on='Twitter')
dfinal=dfinal.merge(Likes,how="inner",on='Twitter')
dfinal=dfinal.merge(verify,how="inner",on='Twitter')
dfinal=dfinal.merge(lists_num,how="inner",on='Twitter')
dfinal=dfinal.merge(dates_creation,how="inner",on='Twitter')

dfinal
```
